## Task 1: NumPy Basics
Create a NumPy array of integers from 1 to 20. Compute the mean, maximum, and standard deviation.

```python
import numpy as np

# Create array from 1 to 20
arr = np.arange(1, 21)
print("Array:", arr)

# Compute statistics
mean_val = np.mean(arr)
max_val = np.max(arr)
std_dev = np.std(arr)

print(f"Mean: {mean_val}")
print(f"Maximum: {max_val}")
print(f"Standard Deviation: {std_dev}")
```

## Task 2: Random Matrix and Boolean Masking
Create a 5Ã—5 random matrix and replace all values less than 0.5 with 0.

```python
# Create a 5x5 random matrix
matrix = np.random.rand(5, 5)
print("Original Matrix:\n", matrix)

# Replace values less than 0.5 with 0 using boolean indexing
matrix[matrix < 0.5] = 0
print("\nModified Matrix (Values < 0.5 replaced with 0):\n", matrix)
```

## Task 3: Pandas Loading and Analysis
Load a CSV file, display the first 5 rows, and calculate the average of one numeric column.

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('/content/sample_data.csv')

# Display first 5 rows
print("First 5 rows:")
display(df.head())

# Calculate average of 'Score' column
avg_score = df['Score'].mean()
print(f"\nAverage Score: {avg_score}")
```

## Task 4: Pandas Filtering and Saving
Filter rows where a column value exceeds a threshold and save the result to a new CSV file.

```python
# Filter rows where Age > 25
filtered_df = df.loc[df['Age'] > 25]
print("Filtered Data (Age > 25):")
display(filtered_df)

# Save to a new CSV file
filtered_df.to_csv('filtered_data.csv', index=False)
print("\nFiltered data saved to 'filtered_data.csv'")
```

## Task 5: Matplotlib Line Plot
Plot a line graph of values from 1 to 10 and label axes properly.

```python
import matplotlib.pyplot as plt

x = range(1, 11)
y = [val**2 for val in x] # Example: y = x^2

plt.plot(x, y)
plt.xlabel('X-Axis (Values 1-10)')
plt.ylabel('Y-Axis (Squared Values)')
plt.title('Line Graph of Squares')
plt.show()
```

## Task 6: Matplotlib Bar Chart
Create a bar chart showing values for 5 categories.

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 24, 18, 30, 22]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Category Comparison')
plt.xticks(categories)
plt.show()
```
