```python
# List to store student records
students = []

# 1ï¸âƒ£ Add Student
def add_student():
    student_id = input("Enter Student ID: ")
    # Check if student already exists
    for s in students:
        if s['id'] == student_id:
            print("âŒ Student ID already exists!\n")
            return

    name = input("Enter Student Name: ")
    department = input("Enter Department: ")
    semester = input("Enter Semester: ")

    student = {
        "id": student_id,
        "name": name,
        "department": department,
        "semester": semester
    }

    students.append(student)
    print("âœ… Student added successfully!\n")


# 2ï¸âƒ£ View All Students
def view_students():
    if not students:
        print("âŒ No students found!\n")
        return

    print("\nğŸ“‹ All Students:")
    for i, student in enumerate(students, start=1):
        print(f"\nStudent {i}")
        print(f"ID: {student['id']}")
        print(f"Name: {student['name']}")
        print(f"Department: {student['department']}")
        print(f"Semester: {student['semester']}")


# 3ï¸âƒ£ Search Student by ID
def search_student():
    student_id = input("Enter Student ID to search: ")
    for student in students:
        if student['id'] == student_id:
            print("\nğŸ” Student Found:")
            print(f"ID: {student['id']}")
            print(f"Name: {student['name']}")
            print(f"Department: {student['department']}")
            print(f"Semester: {student['semester']}\n")
            return
    print("âŒ Student not found!\n")


# 4ï¸âƒ£ Update Student by ID
def update_student():
    student_id = input("Enter Student ID to update: ")
    for student in students:
        if student['id'] == student_id:
            print("Leave blank if no change")
            name = input(f"Enter new Name ({student['name']}): ") or student['name']
            department = input(f"Enter new Department ({student['department']}): ") or student['department']
            semester = input(f"Enter new Semester ({student['semester']}): ") or student['semester']

            student['name'] = name
            student['department'] = department
            student['semester'] = semester

            print("âœ… Student updated successfully!\n")
            return
    print("âŒ Student not found!\n")


# 5ï¸âƒ£ Delete Student by ID
def delete_student():
    student_id = input("Enter Student ID to delete: ")
    for student in students:
        if student['id'] == student_id:
            students.remove(student)
            print("âœ… Student deleted successfully!\n")
            return
    print("âŒ Student not found!\n")


# 6ï¸âƒ£ Main Menu
while True:
    print("===== University Management System =====")
    print("1. Add Student")
    print("2. View All Students")
    print("3. Search Student")
    print("4. Update Student")
    print("5. Delete Student")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        view_students()
    elif choice == "3":
        search_student()
    elif choice == "4":
        update_student()
    elif choice == "5":
        delete_student()
    elif choice == "6":
        print("ğŸ‘‹ Exiting system. Goodbye!")
        break
    else:
        print("âš  Invalid choice, try again.\n")
```
