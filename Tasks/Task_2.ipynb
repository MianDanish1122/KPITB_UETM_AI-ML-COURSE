### TASK 4 — FEATURE CORRELATION (SALES DATA)



```python
import numpy as np

# 1. Create two numerical lists: advertising budget and sales
ads_budget = np.array([100, 200, 300, 400, 500])
sales = np.array([10, 20, 30, 40, 50])

print(f"Advertising Budget: {ads_budget}")
print(f"Sales: {sales}")

# 2. Calculate the correlation between advertising budget and sales
correlation_matrix = np.corrcoef(ads_budget, sales)
correlation_coefficient = correlation_matrix[0, 1]

print(f"\nCorrelation matrix:\n{correlation_matrix}")
print(f"Correlation coefficient between advertising budget and sales: {correlation_coefficient:.2f}")
```

### TASK 5 — MODEL LEARNING SIMULATION



```python
# 1. Initialize a weight (random starting point)
w = 20
# Learning rate (controls the step size)
lr = 0.1

print(f"Initial Weight: {w:.2f}")

# Simulate 10 steps of gradient descent
for i in range(1, 11):
    # Calculate the loss (e.g., squared error for a simple model)
    loss = w**2

    # Calculate the gradient of the loss with respect to the weight (derivative of w^2 is 2w)
    gradient = 2 * w

    # Update the weight using the gradient descent formula
    w = w - lr * gradient

    print(f"Step {i:2d}, Weight: {w:8.4f}, Loss: {loss:8.4f}")
```

### TASK: Linear Regression Model for Sales


```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Ensure the advertising budget and sales data are defined (from Task 4)
# If not, uncomment and run the following lines:
import numpy as np
ads_budget = np.array([100, 200, 300, 400, 500])
sales = np.array([10, 20, 30, 40, 50])

# Reshape the input data for sklearn (needs to be 2D array)
X = ads_budget.reshape(-1, 1) # Features
y = sales # Target

# 1. Create a Linear Regression model
model = LinearRegression()

# 2. Train the model
model.fit(X, y)

# 3. Make predictions
y_pred = model.predict(X)

print(f"Model Intercept: {model.intercept_:.2f}")
print(f"Model Coefficient (slope): {model.coef_[0]:.2f}")

# 4. Visualize the results
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Actual Sales')
plt.plot(X, y_pred, color='red', label='Predicted Sales (Regression Line)')
plt.title('Linear Regression: Sales vs. Advertising Budget')
plt.xlabel('Advertising Budget')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```
