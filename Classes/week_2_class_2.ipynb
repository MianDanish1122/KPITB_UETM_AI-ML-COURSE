```python
import pandas as pd

df_housing = pd.read_csv('/content/Housing.csv')
df_housing.head()
```

```python
print("DataFrame Info:")
df_housing.info()

print("\nMissing values in each column:")
print(df_housing.isnull().sum())

# Separate features (X) and target (Y)
X_housing = df_housing.drop('price', axis=1)
Y_housing = df_housing['price']

# Identify categorical columns
categorical_cols = X_housing.select_dtypes(include=['object']).columns
print(f"\nCategorical columns identified: {list(categorical_cols)}")

# Apply one-hot encoding to categorical columns
X_housing_encoded = pd.get_dummies(X_housing, columns=categorical_cols, drop_first=True)
print("\nShape of features after one-hot encoding:", X_housing_encoded.shape)

# Split the preprocessed data into training and testing sets
X_housing_train, X_housing_test, Y_housing_train, Y_housing_test = train_test_split(X_housing_encoded, Y_housing, test_size=0.2, random_state=42)

print("\nData split into training and testing sets:")
print(f"X_housing_train shape: {X_housing_train.shape}")
print(f"X_housing_test shape: {X_housing_test.shape}")
print(f"Y_housing_train shape: {Y_housing_train.shape}")
print(f"Y_housing_test shape: {Y_housing_test.shape}")
```

```python
from sklearn.linear_model import LinearRegression

# Instantiate a Linear Regression model
housing_linear_model = LinearRegression()

# Fit the model to the training data
housing_linear_model.fit(X_housing_train, Y_housing_train)

print("Linear Regression model trained successfully.")
```

```python
Y_housing_pred = housing_linear_model.predict(X_housing_test)

from sklearn.metrics import mean_absolute_error

mae_housing = mean_absolute_error(Y_housing_test, Y_housing_pred)
print(f"Mean Absolute Error (MAE) for Housing Price Prediction: {mae_housing:.2f}")
```

```python
from sklearn.linear_model import Ridge

# Instantiate a Ridge Regression model with default alpha (1.0)
housing_ridge_model = Ridge(alpha=1.0)

# Fit the model to the training data
housing_ridge_model.fit(X_housing_train, Y_housing_train)

print("Ridge Regression model trained successfully.")
```

```python
from sklearn.linear_model import Lasso

# Instantiate a Lasso Regression model with default alpha (1.0)
housing_lasso_model = Lasso(alpha=1.0)

# Fit the model to the training data
housing_lasso_model.fit(X_housing_train, Y_housing_train)

print("Lasso Regression model trained successfully.")
```

```python
Y_housing_pred_lasso = housing_lasso_model.predict(X_housing_test)
print("Lasso Regression predictions made on the test set.")
```
